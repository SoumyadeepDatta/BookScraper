```python
df.drop_duplicates(subset="Title",keep=False,inplace=True)
df.reset_index(drop=True,inplace=True)
```

```python
df.shape
```

```python
df['Author_new'] = df['Author'].apply(lambda x:x.replace(" ",""))
df['Publisher'] = df['Publisher'].apply(lambda x:x.replace(" ",""))
```

```python
df.head(2)
```

```python
def convert3(text):
    l3=[]
    l3.append(text)
    return l3
```

```python
df['Author_new'] = df['Author_new'].apply(convert3)
df['Publisher'] = df['Publisher'].apply(convert3)
```

```python
df.head(2)
```

```python
df['Blurb_new'] = df['Blurb'].apply(lambda x:x.split())
```

```python
df.head(2)
```

```python
df['tags'] = df['Blurb_new'] + df['Author_new'] + df['Publisher'] 
```

```python
df = df.iloc[:15000,:]
```

```python
df.shape
```

```python
book_df = df.drop(columns=['Publisher','Author_new','Blurb_new'])
```

```python
book_df.head(2)
```

```python
book_df['tags']=book_df['tags'].apply(lambda x:" ".join(x))
```

```python
book_df.head(2)
```

```python
book_df['tags'][0]
```

```python
book_df.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')#stopwords means meaningless words like : it ,is ,to etc
vector=cv.fit_transform(book_df['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
```

```python
cosine_similarity(vector).shape
```

```python
def recommend(books):
    R=[]
    index=book_df[book_df['Title']==books].index[0]
    distances=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
    
    for i in distances[1:21]:
            R.append(book_df.iloc[i[0]]['Title'])
    distances.clear()
    return R
```

```python
book_df['Recommended_Books'] = book_df['Title'].apply(recommend)
book_df.head(5)
```

```python
def get_name(text):
    N=book_df[book_df['Title']==text]['Recommended_Books'].values[0]
    for i in range (6):
        print(N[i])
        print(book_df[book_df['Title']==N[i]]['Author'].values[0])
        print(book_df[book_df['Title']==N[i]]['ISBN'].values[0])
        print(book_df[book_df['Title']==N[i]]['Blurb'].values[0])
        print("\n")
```
